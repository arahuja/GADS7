```python
import pandas as pd
data = pd.read_csv("http://data.princeton.edu/wws509/datasets/salary.dat", sep='\s+')
```

```python
data.head()
```

```python
import statsmodels.formula.api as sm
```

```python
model = sm.ols(" sl ~ sx ", data).fit()
```

```python
model.summary()
```

```python
model = sm.ols(" sl ~ sx - 1 ", data).fit()
model.summary()
```

```python
model = sm.ols(" sl ~ sx*rk + C(yd)", data).fit()
model.summary()
```

```python
print data.groupby(['rk', 'sx']).count()
```

```python
from sklearn import linear_model
from patsy import dmatrices

y, X = dmatrices('sl ~ sx + yr + rk', data=data, return_type='dataframe')
import numpy as np
y = np.array(y).reshape(y.shape[0])

model = linear_model.LinearRegression()

l1 = linear_model.Lasso( alpha=10000 )
#l2 = linear_model.RidgeCV( alphas=[0.01, 0.1, 10])

l1.fit(X,y)

l1.coef_
```

```python
l2 =linear_model.Ridge( alpha = 10000)
l2.fit(X,y)
l2.coef_
```
